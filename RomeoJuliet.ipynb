```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
def RomeoJulietSim(Rn,Jn,Ar,Aj,Pr,Pj):  
    R=[Rn]
    J=[Jn]
    for i in range (100):
        Rn1=(Ar*Rn)+(Pr*Jn)
        Jn1=(Aj*Jn)+(Pj*Rn)
        R.append(Rn1)
        J.append(Jn1)
        Rn=Rn1
        Jn=Jn1
    return(R,J)
```

```python
R,J=RomeoJulietSim(1,1,0.5,0.7,0.2,0.5)
t=[n for n in range (len(R))]
plt.plot(t,R,color='blue')
plt.plot(t,J,color='red')
```
